### **K-Means Tutorial**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

### **載入資料Loading the Data**

```python
df = pd.read_csv('./customer.csv')
# Lable Encoding
dict1 = {'男':1, '女':2}
df['性別'].replace(dict1, inplace=True)
df
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df)
kmeans.labels_
```

```python
new_df = df.copy()
new_df['第幾群'] = kmeans.labels_
new_df
```

```python
df2 = new_df[new_df['第幾群']==2].iloc[0:30,:]
df2
```

```python
new_dy = kmeans.predict(df)
new_dy # 分群動作

kmeans.cluster_centers_
```
